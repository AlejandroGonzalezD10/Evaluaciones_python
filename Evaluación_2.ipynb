````python
#Atributos de istancia

class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model
        self.battery = 100.0  # bateria en %
        self.distance_traveled = 0.0  # km recorridos por el robot
        self.speed = 0.0  # km/h
        self.direction = 0  # grados (0 a 359)
        print("Robot " + self.name + " modelo " + self.model + " creado")

    def __str__(self):
        # Texto que representa el robot
        return self.name + " (" + self.model + ") - Batería: " + str(round(self.battery, 1)) + "% - Distancia: " + "{:.2f}".format(self.distance_traveled) + "km"

    def move(self, speed, direction, time_minutes):
        # Actualizacion de velocidad y dirección
        self.speed = speed
        self.direction = direction

        # Calculo de distancia
        horas = time_minutes / 60.0
        distancia = speed * horas

        # Suma de la distancia
        self.distance_traveled = self.distance_traveled + distancia

        # Consumo de bateria
        consumo = distancia * 10
        self.battery = self.battery - consumo

        # Evitar la batería negativa
        if self.battery < 0:
            self.battery = 0

        # Aviso de batería baja
        if self.battery < 20:
            print("Batería baja!")

    def charge(self, time_minutes):
        # Carga 2% por minuto
        aumento = time_minutes * 2
        self.battery = self.battery + aumento

        # No pasar del 100%
        if self.battery > 100:
            self.battery = 100

        print(self.name + " cargando... Batería: " + str(round(self.battery, 1)) + "%")

    def get_status(self):
        print(self.__str__())


# Parte 2: Herencia

class WorkerRobot(Robot):
    def __init__(self, name, model):
        # Llamamos al constructor de Robot
        super().__init__(name, model)
        # Atributo extra: carga en kg
        self.cargo = 0

    def move(self, speed, direction, time_minutes):
        super().move(speed, direction, time_minutes)

        # Si tiene carga, gasta 20% más batería
        if self.cargo > 0:
            horas = time_minutes / 60.0
            distancia = speed * horas
            extra = (distancia * 10) * 0.20  # aca se gasta el 20% mas
            self.battery = self.battery - extra
            if self.battery < 0:
                self.battery = 0

    def load_cargo(self, weight):
        # Aumenta la carga
        self.cargo = self.cargo + weight
        print("Cargando " + str(weight) + "kg. Total: " + str(self.cargo) + "kg")

    def unload_cargo(self):
        # Quita la carga
        self.cargo = 0
        print("Carga descargada")


class GuardRobot(Robot):
    def __init__(self, name, model):
        # Constructor del robot normal
        super().__init__(name, model)
        # Atributo extra: modo alerta
        self.alert_mode = False

    def charge(self, time_minutes):
        # En modo alerta carga 50% más lento
        if self.alert_mode == True:
            self.battery = self.battery + (time_minutes * 1)
        else:
            self.battery = self.battery + (time_minutes * 2)

        if self.battery > 100:
            self.battery = 100

        print(self.name + " cargando... Batería: " + str(round(self.battery, 1)) + "%")

    def toggle_alert(self):
        # Cambiar modo alerta
        if self.alert_mode == False:
            self.alert_mode = True
        else:
            self.alert_mode = False

        if self.alert_mode == True:
            print("Modo alerta: ON")
        else:
            print("Modo alerta: OFF")


# Parte 3: Composición


class RobotTeam:
    def __init__(self, team_name):
        self.team_name = team_name
        self.robots = []  # lista de robots
        self.missions_completed = 0  # contador de las misiones de los robots
        print("Equipo " + self.team_name + " formado")

    def add_robot(self, robot):
        # Agregar un robot a la lista
        self.robots.append(robot)
        print(robot.name + " se unió al equipo " + self.team_name)

    def __len__(self):
        # Cantidad de robots que hay
        return len(self.robots)

    def charge_all(self, time_minutes):
        # Cargar los robots en el equipo
        for r in self.robots:
            r.charge(time_minutes)

    def team_status(self):
        # Mostrar estado de todos
        print("\n--- Estado del equipo " + self.team_name + " ---")
        for r in self.robots:
            r.get_status()
        print("Total robots: " + str(len(self)) + " - Misiones completadas: " + str(self.missions_completed))

    def complete_mission(self):
        # Sumar una misión
        self.missions_completed = self.missions_completed + 1
        print("Misión completada Total: " + str(self.missions_completed))


# Parte 4: Código de Prueba


if __name__ == "__main__":
    # 1) Crear robots
    r1 = Robot("Alpha", "RX-1")
    r2 = Robot("Beta", "ZX-2")
    w1 = WorkerRobot("CargoBot", "WR-5")
    g1 = GuardRobot("Sentinel", "GR-9")

    # 2) Probar métodos básicos
    r1.move(10, 90, 30)     # 10 km/h - 30 min
    r2.move(20, 180, 15)    # 20 km/h - 15 min
    w1.load_cargo(50)       # cargar 50 kg
    w1.move(15, 45, 20)     # mover con carga
    g1.toggle_alert()       # activar alerta
    g1.move(5, 270, 60)

    # Cargar cada robot
    r1.charge(20)
    r2.charge(10)
    w1.charge(30)
    g1.charge(15)

    # Mostrar estados
    r1.get_status()
    r2.get_status()
    w1.get_status()
    g1.get_status()

    # 3) Crear equipo
    team = RobotTeam("Titanes")
    team.add_robot(r1)
    team.add_robot(r2)
    team.add_robot(w1)
    team.add_robot(g1)
    print("Número de robots en el equipo: " + str(len(team)))

    # Cargar todo el equipo
    team.charge_all(10)

    # Estado del equipo
    team.team_status()

    # Completar una misión
    team.complete_mission()
````
